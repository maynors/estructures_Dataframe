### Descripción de variables y justificación por que se eliminan o mantienen

>**'Unnamed: 0'**,		¿Por qué se elimina?	Tiene la forma de un correlativo de los registros y algunos números no llevan un orden especifico. Al trabajar con pandas de igual forma existe un index, así que se duplica la columna, la ventaja del index de pandas es que mantiene un orden.

> **Year**,	Año, en este caso todos los datos son del año 2008.	¿Por qué se elimina?	Se unifica en una sola columna (Año - mes - día )

>**Month**,	Meses en numeros (1 - 12).	¿Por qué se elimina?	Se unifica en una sola columna (Año - mes - día )

>**DayofMonth'**,	Dia del mes (1-31).	¿Por qué se elimina?	Se unifica en una sola columna (Año - mes - día )

>**DayOfWeek'**,	Dia de la semana (1 - 7 ).	¿Por qué se mantiene?	Puede ser interesante el análisis de los días con más retrasos por aerolínea, aeropuerto.

>**DepTime'**,	Hora de despegue del avion real. Hora en formato militar.	¿Por qué se mantiene?	Se puede analizar cuales son los horarios en los que más hay retrasos.

>**CRSDepTime**,	Hora de despegue del avion planificada o esperada. Hora en formato militar.	¿Por qué se mantiene?	Se mantiene como referencia y futuros análisis. Por ejemplo se podría decir conclusiones como: "El vuelo de las 11pm generalmente tiene retrasos".

>**ArrTime'**,	Hora de aterrizaje del avion real. Hora en formato militar.	¿Por qué se mantiene?	Se puede analizar cuales son los horarios en los que más hay retrasos al momento de arribar. 

>**CRSArrTime**',	Hora de aterrizaje del avion planificada o esperada. Hora en formato militar.	¿Por qué se mantiene?	Se mantiene como referencia y futuros análisis. Por ejemplo se podría decir conclusiones como: "El vuelo que arriba a las 3 am generalmente tiene retrasos".

>**UniqueCarrier'**,	Código unico de la aerolínea.	¿Por qué se mantiene?	Importante para identificar las aerolíneas con mas retrasos.

>**FlightNum'**,	Número de vuelo.	¿Por qué se mantiene?	Se puede identificar que vuelos son los que tienen más retrasos.

>**TailNum'**,	Es el identificador único de los aviones. Cada avión tiene un código generalmente entre 2-6 caracteres. Aunque el nombre indica que debería estar impreso en la "cola" del avión no siempre es así, puede ser impreso en otras partes. Con este identificador puedes encontrar información como el mantenimiento que ha tenido, los lugares que ha viajado, etc.	¿Por qué se mantiene?	Se puede identificar que aviones causan más retrasos.

>**ActualElapsedTime'**,	Tiempo real del avión para despegar y llegar al destino. (minutos).	¿Por qué se mantiene?	Resta entre DepTime y ArrTime. Se puede analizar en que vuelos (corta duración o larga) son más propensas los retrasos.

>**CRSElapsedTime'**,	Tiempo planificado/esperado del avión para despegar y llegar al destino. (minutos).	¿Por qué se mantiene?	Resta entre CRSDepTime y CRSArrTime. Sirve como referencia, por ejemplo, se pueden concluir cosas como "en los vuelos planificados de más de 4 horas tienen a existir más retrasos".

>**AirTime'**,	Tiempo que pasa el avión en el aire. (minutos).	¿Por qué se mantiene?	Es necesario para hacer calculos de velocidad media si se combina con la distancia entre cada aeropuerto.

>**ArrDelay**',	Minutos de retraso al momento de aterrizar.	¿Por qué se mantiene?	Es la resta del tiempo de salida y llegada del avión. En este dato se basa el análisis.

>**DepDelay'**,	Minutos de retraso en el despegue..	¿Por qué se mantiene?	Es la resta entre DepTime y CRSDepTime. Columna útil para los calculos y analizar en que aeropuertos se producen más los retrasos en el despegue.

>**Origin'**,	Código del aeropuerto de origen (IATA= International Air Transport Association).	¿Por qué se mantiene?	Se puede analizar en que aeropuertos se originan muchos retrasos en la salida.

>**>Dest'**,	Código del aeropuerto de destino (IATA= International Air Transport Association).	¿Por qué se mantiene?	Se puede analizar en que aeropuertos se originan muchos retrasos en la llegada.
Dest',	Código del aeropuerto de destino (IATA= International Air Transport Association).	¿Por qué se mantiene?	Se puede analizar en que aeropuertos se originan muchos retrasos en la llegada.

>**Distance'**,	Distancia entre el origen y el destino en millas.	¿Por qué se mantiene?	Es necesario para hacer calculos de velocidad media si se combina con el tiempo en el aire del avión.

>**TaxiIn'**,	Tiempo de entre el aterrizaje del avión y cuando se parquea "activando" los frenos.	¿Por qué se mantiene?	Interesante para analizar el tiempo antes del despegue.

>**TaxiOut'**,	Teimpo desde que "desactiva" los frenos desde su posición de parque y cuando despega el avión.	¿Por qué se mantiene?	Interesante para analizar el tiempo que se tarda el avión despues del aterrizaje. Por ejemplo, puede ser que el avión arribó en tiempo pero no hay suficientes "gates" para aparcar.

>**Cancelled**',	Indica si el vuelo ha sido cancelado o no. 0 = No, 1=Si.	¿Por qué se mantiene?	Se puede analizar que aerolínea cancela más vuelos.

>**CancellationCode'**,	Código que indica la razón de la cancelación del vuelo (A=aerolínea, B=clima, c=NAS, D=seguridad). **NAS: Retrasos y cancelaciones atribuibles al sistema de aviación nacional que se refieren a un amplio conjunto de condiciones, tales como condiciones climáticas no extremas, operaciones aeroportuarias, volumen de tráfico pesado y control del tráfico aéreo.	¿Por qué se mantiene?	Se puede analizar por que se cancelan la mayor cantidad de vuelos en el año

>**Diverted'**,	Indica si el vuelo ha sido desviado de su trayectoria original. 1 = SI, 0 = NO.	¿Por qué se mantiene?	Se puede analizar en que rutas generalmente existen más desviaciones. 

>**CarrierDelay**	Retrasos debido a la aerolínea en minutos.	¿Por qué se mantiene?	Se puede analizar el % del tiempo de retraso asociado a la variable.

>**LateAircraftDelay**	Retrasos debido a que el mismo avión llego tarde y en consecuencia el siguiente vuelo tiene retraso.	¿Por qué se mantiene?	Se puede analizar el % del tiempo de retraso asociado a la variable.

>**WeatherDelay**	Retrasos debido al clima en minutos.	¿Por qué se mantiene?	Se puede analizar el % del tiempo de retraso asociado a la variable.

>**NASDelay**	Retrasos debido a la NAS en minutos.	¿Por qué se mantiene?	Se puede analizar el % del tiempo de retraso asociado a la variable.

>**SecurityDelay**	Retrsos debido a las seguridad en minutos.	¿Por qué se mantiene?	Se puede analizar el % del tiempo de retraso asociado a la variable.

### Análizando el ArrDelay con las otras variables. 

Una variable muy importante y crítica es la que indica el retraso en el arribo. Por lo cual, a continuación se comparara con las otras variables para ver en que porcentaje le afecta y esta relacionada. Para eso se realiza un matriz de correlación

```python
dflights_corr= dflights[[
                #'DayOfWeek', 
                #'DepTime', 
                #'CRSDepTime', 
                #'ArrTime', 
                #'CRSArrTime',
                #'UniqueCarrier',
                #'FlightNum', 
                #'TailNum', 
                #'ActualElapsedTime',
                'ArrDelay',
                'CRSElapsedTime', 
                #'AirTime', 
                'DepDelay', 
                #'Origin', 
                #'Dest',
                'Distance', 
                'TaxiIn', 
                'TaxiOut', 
                #'Cancelled', 
                #'CancellationCode',
                #'Diverted', 
                'CarrierDelay', 
                'WeatherDelay', 
                'NASDelay', 
                'SecurityDelay',
                'LateAircraftDelay']].copy()
dflights_corr.corr()
```

### Observaciones de la matriz de correlaciones

> Se puede observar que no existe una correlación entre la distancia y el tiempo de retraso. Es decir, no esta relacionado que un vuelo por ser más largo tiene más probabilidad de en arribar con retraso.

> Lo que era lógico imaginar es que si hay retraso en el despegue también llegara con retraso al destino en la mayoría de ocasiones.

> Un análisis cruzado puede surgir de analizar el Carrier delay vs { ArrDelay (0.51 correlacion) y DepDelay(0.54 correlacion)} , es decir que el Carrier delay afecta más en el retraso asociado al despegue que al del arribo.

>En el caso del WeatherDelay vs { ArrDelay(0.26 correlacion) y DepDelay (0.24 correlacion) }, el caso es contrario afecta más en el arribo que en el despegue. 

>El NAS delay afecta más en los arribos que en el despegue. Por ejemplo, si no hay espacio en pista para que te aterrice le pediran que este en el aire más tiempo.

## Crea columnes noves (velocitat mitjana del vol, si ha arribat tard o no...)

```python
#Agregando una columna con la velocidad media en el aire (millas / horas)
dflights['VelMediaVuelo'] = (dflights['Distance'] / ((dflights['AirTime'])/60) )
```

```python
# Se agrega una columna si arriba tarde o no, 1=Tarde, 0=en tiempo, se considera tarde si arriba después de 15 minutos. 
dflights['EstadoVuelollegada'] = ["Tarde" if x > 15 else "En tiempo" for x in dflights['ArrDelay']]
```

```python
#Agrego una columna con los nombres de las aerolíneas
airlines= pd.read_csv('https://raw.githubusercontent.com/beanumber/airlines/master/data-raw/airlines.csv')
airlines = airlines.rename(columns ={"Code": "UniqueCarrier"})
dflights = dflights.merge(airlines, how='left', on='UniqueCarrier')
```

```python
#Número total de vuelos por aerolínea
dflights['Description'].value_counts()
```

```python
#Cantidad de vuelos con llegada tarde
dflights_tarde= dflights.query("EstadoVuelollegada=='Tarde'")
dflights_tarde["Description"].value_counts().sort_values( ascending=False)
```

```python
#Quins són els vols més llargs?
dflights[['FlightNum','Origin','Dest','Distance',]].sort_values('Distance',ascending=False).drop_duplicates()
```

```python
#Quins són els més endarrerits??
dflights[['FlightNum','Origin','Dest','Distance','ArrDelay']].sort_values('ArrDelay',ascending=False).drop_duplicates().head(15)
```

```python
#Es interesante calcular los días donde hay mas retrasos
# El día domingo es donde se pueden observar más retrasos.
dflights.groupby('DayOfWeek')['ArrDelay'].mean().sort_values(ascending=False)
```

```python
#Retrasos en día domingo
dflights.groupby('DayOfWeek')['ArrDelay'].mean().sort_values(ascending=False)
```

```python
#Aerolíneas con más retrasos en promedio, se elige CarrierDelay 
dflights.groupby('Description')['CarrierDelay'].mean().sort_values(ascending=False)
```
