### Análizando el ArrDelay con las otras variables. 

Una variable muy importante y crítica es la que indica el retraso en el arribo. Por lo cual, a continuación se comparara con las otras variables para ver en que porcentaje le afecta y esta relacionada. Para eso se realiza un matriz de correlación y un heatmap.

```python
correlaciones = dflights [['ArrDelay',
                        'DepDelay',
                        'AirTime',
                        'Distance',
                        'CarrierDelay',
                        'WeatherDelay',2
                        'NASDelay',
                        'SecurityDelay',
                        'LateAircraftDelay']].corr()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Creamos el heat map
fig, ax = plt.subplots(figsize=(10,5)) 
sns.heatmap(correlaciones, annot=True, linewidths=.5, cmap="coolwarm", linecolor='black')

# Mostramos el gráfico
plt.show()
```

### Observaciones de la matriz de correlaciones

> Se puede observar que no existe una correlación entre la distancia o el tiempo que pasa en el aire con el tiempo de retraso. Es decir, no esta relacionado que un vuelo por ser más largo tiene más probabilidad de en arribar con retraso.

> Lo que era lógico imaginar es que si hay retraso en el despegue también llegara con retraso al destino en la mayoría de ocasiones. ArrDelay vs DepDelay(0.95 correlación).

> Un análisis cruzado puede surgir de analizar el Carrier delay vs { ArrDelay (0.53 correlacion) y DepDelay(0.56 correlacion)} , es decir que el Carrier delay afecta más en el retraso asociado al despegue que al del arribo. Aunque no son valores tan alejados para ser concluyentes.

> La causa que más afecta el retraso de los aviones es "LateAircraftDelay" con 0.56 de correlación, seguido por el CarrierDelay con 0.53 y en tercer lugar el NASDelay (0.46).


## Crea columnes noves (velocitat mitjana del vol, si ha arribat tard o no...)

```python
#Agregando una columna con la velocidad media en el aire (millas / horas)
dflights['VelMediaVuelo'] = (dflights['Distance'] / ((dflights['AirTime'])/60) )
```

```python
# Se agrega una columna si arriba tarde o no, 1=Tarde, 0=en tiempo, se considera tarde si arriba después de 15 minutos. 
dflights['EstadoVuelollegada'] = ["Tarde" if x > 15 else "En tiempo" for x in dflights['ArrDelay']]
```

```python
# Se agrega una columna si arriba tarde o no, 1=Tarde, 0=en tiempo, se considera tarde si arriba después de 15 minutos. 
dflights['BinarioTardeOnTime'] = [1 if x > 15 else 0 for x in dflights['ArrDelay']]
```

```python
#Agrego una columna con los nombres de las aerolíneas
airlines= pd.read_csv('https://raw.githubusercontent.com/beanumber/airlines/master/data-raw/airlines.csv')
airlines = airlines.rename(columns ={"Code": "UniqueCarrier"})
dflights = dflights.merge(airlines, how='left', on='UniqueCarrier')
```
